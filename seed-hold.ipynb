```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
import matplotlib.dates as mdates
from pycoingecko import CoinGeckoAPI
import requests
import matplotlib.ticker as ticker
```

```python
#Cargamos la data
all_tx = pd.read_csv('csv/gq_all_tx.csv')
saved_wallets_read = pd.read_csv('csv/gq_saved_wallets.csv')
```

```python
saved_wallets = saved_wallets_read.query("type == 'venture'")
saved_wallets
```

calcular % de wallets activas que hayan participado en seed y que no

```python
tx_from_venture = all_tx[all_tx['from'].isin(saved_wallets['wallet'])]
tx_from_venture.reset_index(drop=True, inplace=True)
tx_from_venture
```

```python
seed_investors_wallets = pd.DataFrame(tx_from_venture['to'])
seed_investors_wallets.rename(columns={'to': 'wallet'}, inplace=True)
seed_investors_wallets
```

```python
tx_from_seed_investors = all_tx[all_tx['from'].isin(seed_investors_wallets['wallet'])]
tx_from_seed_investors.reset_index(drop=True, inplace=True)
tx_from_seed_investors
```

```python
tx_from_seed_investors = tx_from_seed_investors[['from', 'to', 'value']].copy()
tx_from_seed_investors.rename(columns={'from': 'wallet', 'value': 'sell-stake'}, inplace=True)
tx_from_seed_investors.dropna(inplace=True)
tx_from_seed_investors
```

```python
saved_wallets_stake = saved_wallets_read.query("type == 'stake'")
saved_wallets_stake 
```

```python
seed_investors_stake = tx_from_seed_investors[tx_from_seed_investors['to'].isin(saved_wallets_stake['wallet'])].copy()
seed_investors_stake['stake'] = seed_investors_stake['sell-stake']
seed_investors_stake.drop(columns=['sell-stake'], inplace=True)
seed_investors_stake.reset_index(drop=True, inplace=True)
seed_investors_stake
```

```python
seed_investors_sell = tx_from_seed_investors[~tx_from_seed_investors['to'].isin(saved_wallets_stake['wallet'])].copy()
seed_investors_sell['sell'] = -tx_from_seed_investors['sell-stake']
seed_investors_sell.drop(columns=['sell-stake'], inplace=True)
seed_investors_sell.reset_index(drop=True, inplace=True)
seed_investors_sell
```

```python
tx_to_seed_investors = all_tx[all_tx['to'].isin(seed_investors_wallets['wallet'])]
tx_to_seed_investors.reset_index(drop=True, inplace=True)
tx_to_seed_investors
```

```python
tx_to_seed_investors = tx_to_seed_investors[['from', 'to', 'value']].copy()
tx_to_seed_investors.rename(columns={'to': 'wallet', 'value': 'buy-claim'}, inplace=True)
tx_to_seed_investors.dropna(inplace=True)
tx_to_seed_investors
```

```python
seed_investors_unstake = tx_to_seed_investors[tx_to_seed_investors['from'].isin(saved_wallets_stake['wallet'])].copy()
seed_investors_unstake['unstake'] = -seed_investors_unstake['buy-claim']
seed_investors_unstake.drop(columns=['buy-claim'], inplace=True)
seed_investors_unstake.reset_index(drop=True, inplace=True)
seed_investors_unstake
```

```python
seed_investors_buy = tx_to_seed_investors[~tx_to_seed_investors['from'].isin(saved_wallets_stake['wallet'])& ~tx_to_seed_investors['from'].isin(saved_wallets['wallet'])].copy()
seed_investors_buy['buy'] = tx_to_seed_investors['buy-claim']
seed_investors_buy.drop(columns=['buy-claim'], inplace=True)
seed_investors_buy.reset_index(drop=True, inplace=True)
seed_investors_buy
```

```python
seed_investors_claim = tx_to_seed_investors[~tx_to_seed_investors['from'].isin(saved_wallets_stake['wallet']) & tx_to_seed_investors['from'].isin(saved_wallets['wallet'])].copy()
seed_investors_claim['claim'] = tx_to_seed_investors['buy-claim']
seed_investors_claim.drop(columns=['buy-claim'], inplace=True)
seed_investors_claim.reset_index(drop=True, inplace=True)
seed_investors_claim
```

```python
seed_investors_buy_group = seed_investors_buy.groupby('wallet').sum()
seed_investors_unstake_group  = seed_investors_unstake.groupby('wallet').sum()
seed_investors_sell_group = seed_investors_sell.groupby('wallet').sum()
seed_investors_stake_group = seed_investors_stake.groupby('wallet').sum()
seed_investors_claim = seed_investors_claim.groupby('wallet').sum()
```

```python
merged_seed_investors = pd.concat([seed_investors_buy_group, seed_investors_unstake_group, seed_investors_sell_group, seed_investors_stake_group, seed_investors_claim], axis=1, sort=False)
merged_seed_investors = merged_seed_investors.groupby('wallet')
merged_seed_investors.head()
```
